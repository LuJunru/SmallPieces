```python
import pylab as pl
%pylab inline
```

```python
pl.plot([0,1,2],[0,1,2])
pl.xlabel("some x", fontsize=20)
```

```python
pl.rcParams
```

```python
s = {'font.size': 20.0,
     'figure.figsize': [6.0, 6.0]}
pl.rcParams.update(s)
pl.plot([0,1,2],[0,1,2],'o',color='r')
pl.xlabel("some x", fontsize=20)
```

```python
fig = pl.figure(figsize=(6,8), facecolor='gray')  # 6,8是fig的尺寸
# fig is a container for plot, plot is a container for axes
ax1 = fig.add_subplot(121)  # one row 2 colomun 1st
ax2 = fig.add_subplot(122)
```

```python
x = np.arange(100)
y = np.random.randn(100) + np.sqrt(x)
```

```python
fig = pl.figure(figsize=(13, 4))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.plot(x, y, label="my sq root")
ax1.legend(fontsize=15)
ax2.plot(x, y, '.-')
```

```python
x = np.arange(100)
y = np.random.randn(100) + np.sqrt(x)
yerr = np.sqrt(y)  # create an uncertantiy
```

```python
fig = pl.figure(figsize=(13, 4))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.plot(x, y, '.', label="my sq root")
ax1.legend(fontsize=15)
# ax2.errorbar(x[:10], y[:10], yerr=yerr[:10], fmt='.')
# ax2.errorbar(x, y, yerr=yerr, fmt='ro')  # defalut的参数必须写在前面，带有参数名的参数写在后面
ax2.fill_between(x, y+yerr, y-yerr, alpha=0.3)
```
